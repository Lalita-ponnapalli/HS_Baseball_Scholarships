```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder

import pandas as pd
import tensorflow as tf
```

```python
# Import our input dataset
url = 'https://raw.githubusercontent.com/Lalita-ponnapalli/HS_Baseball_Scholarships/main/allstudentactivities1718final.csv'
schdist_df = pd.read_csv(url)
schdist_df.head()
```

```python
schdist_df.count()
```

```python
schdist_df = schdist_df.drop(columns=['year', 'schoolname'])
```

```python
# Generate our categorical variable list
schdist_cat = schdist_df.dtypes[schdist_df.dtypes == "object"].index.tolist()
schdist_cat
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(schdist_df[schdist_cat]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(schdist_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
schdist_df = schdist_df.merge(encode_df,left_index=True, right_index=True)
schdist_df = schdist_df.drop(schdist_cat,1)
schdist_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = schdist_df["studentcount"].values
X = schdist_df.drop(["studentcount"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
y_train
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="relu"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="mean_absolute_error", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
